2.3 สร้างคอลัมภ์ใหม่ชื่อว่า rounded_stars โดยที่นำคะแนนใน Stars เป็นชนิด float64

```
quiz['rounded_stars'] = quiz['Stars']
```

```
# ข้อนีี้ควรเปลี่ยน unrated เป็น -1 (ที่จริงควร drop ออกเพราะมีผลแต่ตอนนี้ทำวิธีนี้ไปก่อน)

quiz10 = quiz
quiz10['rounded_stars'] = quiz10.rounded_stars.str.replace('Unrated', '-1').astype(float)
quiz10['rounded_stars'] = quiz10['rounded_stars'].astype(np.float64)
quiz10.info()
```
