**1.1 มีจำนวนกี่บริษัท (Company) ที่เปิดรับพนักงาน และมีตำแหน่งรวมกันกี่ตำแหน่ง (นับรวมทุกตำแหน่งตาม Title)**





```
list_company = []
list_title = [] 

for index, row in dataset1.iterrows():
  company = row['Company']
  title = row['Title'].split(',')

  for x in title:
    list_company.append(company)
    list_title.append(x)

dict_ = {"Company": list_company, 
         "Title": list_title}

df1 = pd.DataFrame(dict_)
df1.groupby(['Company', 'Title'])['Title'].agg(['count'])

```

**1.2 ในประเทศ Singapore มีกลุ่มของตำแหน่ง (Category) ที่เปิดรับพนักงาน อะไรบ้าง และมีจำนวนกี่ตำแหน่ง**

```
dataset1[dataset1['Location'] == 'Singapore'].groupby('Category').count()
```

**1.3 รายชื่อตำแหน่งงาน (Title) ที่มีจำเป็นตัองมีคุณสมบัติด้านวุฒิการศึกษา (Minimun Qualifications) อย่างน้อยปริญญาโท (Master Degree)**

```
list_company = []
list_title = [] 
list_minimum = []

for index, row in dataset1.iterrows():
  company = row['Company']
  minimum = row['Minimum Qualifications']
  title = row['Title'].split(',')

  for x in title:
    list_company.append(company)
    list_minimum.append(minimum)
    list_title.append(x)

dict_ = {"Company": list_company, 
         "Title": list_title,
         "Minimum Qualifications": list_minimum}

df1 = pd.DataFrame(dict_)
df1[df1['Minimum Qualifications'].str.contains('Master', na=False)]

```


**1.4 แสดงรายชื่อประเทศ (Location) ที่ประกาศรับสมัครพนักงานมากที่สุด 5 อันดับจากมากไปน้อย**  


```
list_company = []
list_title = [] 
list_Location = []

for index, row in dataset1.iterrows():
  company = row['Company']
  Location = row['Location'].split(',')
  title = row['Title'].split(',')

  for x in title:

    for y in Location:

      list_company.append(company)
      list_Location.append(y)
      list_title.append(x)

dict_ = {"Company": list_company, 
         "Title": list_title,
         "Location": list_Location}

df1 = pd.DataFrame(dict_)
df1.groupby('Location').count().sort_values(by='Company', ascending=False).head(5)

```

**1.5 ความถี่สะสมของคำ (word) ที่เกิดขึ้นในคุณสมบัติเฉพาะตำแหน่ง (Preferred Qualification) สำหรับในรายการตำแหน่ง (Title) ที่มีคำว่า "Engineering" มีจำนวนเท่าไร**

    1.5.1 แสดงผลเป็นกราฟแท่ง เรียงลำดับจากมากไปน้อย 
    1.5.2 การแสดงผลในข้อ 1.5.1 จะไม่รวมความถี่สะสม ของคำต่อไปนี้ "a", "an", "the", "and", "or", "of", "with", "to", "is",  "am", "are", "as", "in" 
    1.5.3 ไม่รวมเครื่องหมายพิเศษเช่น \n (new line) , "." (dot), "," (comma) 

```
dataset1['Preferred Qualifications'] = dataset1['Preferred Qualifications'].astype('str') 
dataset1 = dataset1[dataset1['Title'].str.contains('Engineering', na=False)]

list_company = []
list_title = []
list_Preferred = []

for index, row in dataset1.iterrows():
  company = row['Company']
  title = row['Title'].split(',')
  Preferred = row['Preferred Qualifications'].split()

  for x in title:
    for y in Preferred:
      list_company.append(company)
      list_title.append(x)
      list_Preferred.append(y)

dict_ = {"Company": list_company, 
         "Title": list_title,
         "Preferred": list_Preferred}

df1 = pd.DataFrame(dict_)
# df1 = df1[df1['Title'].str.contains('Engineering', na=False)]
df1["Indexes"]= df1["Title"].str.find('Engineering') 
df1 = df1[df1['Indexes'] == 1]
# df1 = df1.groupby(['Preferred'])['Preferred'].count()
```

**2.5 แสดงกราฟของรายวิชา (Course) ที่มีผู้เข้าเรียน (course_students_enrolled)  มากที่สุด 5 อันดับแรก**
- ให้แกน x แทนจำนวนคน (หน่วยเป็น ล้านคน) แกน y เป็นชื่อรายวิชา

```
dataset2.head(2)
```

```
df2 = dataset2[dataset2['course_Certificate_type'] == 'COURSE']

df2.course_students_enrolled = (df2.course_students_enrolled.replace(r'[km]+$', '', regex=True).astype(float) * \
          df2.course_students_enrolled.str.extract(r'[\d\.]+([km]+)', expand=False)
          .fillna(1)
          .replace(['k','m'], [10**3, 10**6]).astype(int))

df2 = df2.sort_values(by='course_students_enrolled', ascending=False).head(5)
df2
```

**3.1 จากข้อมูลที่ให้มา ให้แสดงรายชื่อประเทศที่การเกิดเหตุการณ์ Tsunami และจำนวนครั้งที่เกิด**

```
df3 = dataset3[dataset3['EVENT_VALIDITY'].str.contains('Tsunami', na=False)]
df3.groupby('COUNTRY')['COUNTRY'].count()
```

**3.2 สาเหตุ (Cause) ของการเกิด Tsunami โดยนับแต่ตั้งแต่ปี 1800 จนถึงปัจจุบัน มีอะไรบ้าง และมีกี่ครั้งที่เกิด ให้แสดงข้อมูลเรียงจากมากไปน้อย**

```
df3 = dataset3[(dataset3['EVENT_VALIDITY'].str.contains('Tsunami', na=False)) & (dataset3['YEAR'] >= 1800) & (dataset3['CAUSE'] != 'Unknown')]
df3.groupby('CAUSE').count().sort_values(by='ID', ascending=False)['ID']
```
